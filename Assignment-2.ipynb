# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
sub1 = int(input("Enter your marks: "))
sub2 = int(input("Enter your marks: "))
sub3 = int(input("Enter your marks: "))
sub4 = int(input("Enter your marks: "))
sub5 = int(input("Enter your marks: "))


obtain_marks = sub1+sub2+sub3+sub4+sub5
pec = (obtain_marks/500)*100
gread = ""
if pec >90:
    gread = "A"
elif pec >80:
    gread = "B"
elif pec >70:
    gread = "C"
elif pec >60:
    gread = "D"
else:
    gread = "Fail"


print(f"""
     total marks :   500
     Obtain marks:   {obtain_marks}
     percentage  :   {pec}
     Grade       :   {gread}""")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter the number of month to find the weather"))

if month == 12 | month <= 2:
    print (f"""Current weather of month {month} is Winter""")

elif month <= 5:
    print (f"""Current weather of month {month} is Spring""")

elif month <= 8:
    print (f"""Current weather of month {month} is Summer""")
    
elif month <= 11:
    print (f"""Current weather of month {month} is Autumn""")
elif month == 0:
    print ("You are out of Month")
else:
    print ("You are out of Month")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = int(input("enter a number"))
num2 = int(input("enter a number"))
operator = (input("enter a operater form these:+, -, *, /"))
ans = ""
if operator =="+":
    ans = num1 + num2
elif operator =="-":
    ans = num1 - num2
elif operator =="*":
    ans = num1 * num2
elif operator =="/":
    ans = num1 / num2
else:
    print("invalid operator")

print(f"""
      first number:   {num1}
      second number:  {num2}
      operator     :  {operator}
      ans          :  {ans}""")
```
